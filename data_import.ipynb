# Separate oct_data and adding them to main_data

```python
#Load positions of oct_data where the time value is 0 in df_start (if time = 0 a new id starts)

df_start = list(np.where(oct_data[:,0] == 0))[0] #positions of the data 0 in the first column of oct_data
df_start
```

```python
#Check types and sizes of different variables

#type(np.where)
#type(df_start)
#type(np.where(oct_data[:,0] == 0))
#np.size(df_start)
#len(df_start)
```

```python
oct_df_list = [] #Init list for oct data

for idx, row in main_data.iterrows(): #Iterate for idx


    if idx < len(main_data)-1: #For lines 1 to 629 of main_data

        oct_df = pd.DataFrame(oct_data[df_start[idx]:df_start[idx+1]], columns = ["time", "depth_measured"]) #Write time and depth_measured for the current id in oct_df

        oct_df.index = oct_df["time"] #Overtake time as index of oct_df

        oct_df = oct_df.drop("time", axis = 1) #Drop collumn time of oct_df

        oct_df_list.append(oct_df) #Append oct_df_list by oct_df of the current iteration
       
    else: #For line 630 of main_data as df_start ends at line 629
    
        oct_df = pd.DataFrame(oct_data[df_start[idx]:-1], columns = ["time", "depth_measured"])

        oct_df.index = oct_df["time"]

        oct_df = oct_df.drop("time", axis = 1)

        oct_df_list.append(oct_df)   
```

```python
#Print oct_df_list
oct_df_list
```

```python
#Check type and size of some variables

#type(oct_df_list)
#type(main_data)
#len(oct_df_list)
```

```python
#Add oct_df_list to main_data

main_data["oct_data"] = oct_df_list
```

```python
#Print main_data including oct_data

main_data
```

# Modify main_data

```python
#Set id as index in main_data and drop column id

main_data.index = main_data["id"]

main_data = main_data.drop("id", axis = 1)
```

```python
#Change unit of depth in Âµm

main_data.depth = main_data.depth * 1000
```

```python
#Drop t_inters as it is no longer needed according to Readme.txt

main_data = main_data.drop("t_inters", axis=1)
```
